# SCATTER PLOT

```python
df= px.data.iris()
df.head()
```

```python
df['species'].value_counts()
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color='species', size='petal_length', height=600)
fig.show()
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color='petal_length', size='petal_length', height=600)
fig.show()
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color='species', symbol='species', height=600)
fig.show()
```

```python
df_t= px.data.tips()
df_t.head()
```

```python
fig = px.scatter(df_t, x='total_bill', y='tip',
                color='smoker', facet_col='sex',facet_row='time')
fig.show()
```

```python
fig = px.scatter(df, x="sepal_length", y="sepal_width",
                marginal_x='histogram',
                marginal_y='rug')
fig.show()
```

```python
fig = px.scatter(df_t, x='total_bill', y='tip',
                trendline='ols')
fig.show()
```

# Histogram

```python
fig = px.histogram(df_t, x='total_bill')
fig.show()
```

```python
fig = px.histogram(df_t, x='day')
fig.show()
```

```python
fig = px.histogram(df_t, x='total_bill', nbins=50)
fig.show()
```

```python
stocks = px.data.stocks()
stocks.head()
```

```python
fig = px.histogram(stocks, x='date')
fig.update_layout(bargap=0.5)
fig.show()
```

```python
fig = px.histogram(df_t, x='total_bill', histnorm= 'probability density')
fig.show()
```

```python
fig = px.histogram(df_t, x='total_bill', color= 'sex')
fig.show()
```

```python
fig = px.histogram(df_t, x='total_bill',y= 'tip', histfunc= 'avg')
fig.show()
```

```python
fig = px.histogram(df_t, x='sex', y='total_bill', color='sex',
                  pattern_shape='smoker')
fig.show()
```

```python
fig = px.histogram(df_t, x='total_bill',y= 'tip', histfunc= 'avg',
                  nbins=8, text_auto=True)
fig.show()
```

# BoxPlot

```python
df_t.head()
```

```python
fig= px.box(df_t, y='total_bill')
fig.show()
```

```python
fig= px.box(df_t, x='time', y='total_bill')
fig.show()
```

```python
fig= px.box(df_t, x= 'time', y='total_bill', points='all')
fig.show()
```

```python
fig= px.box(df_t, x= 'time', y='total_bill', points='outliers')
fig.show()
```

```python
fig= px.box(df_t, x= 'time', y='total_bill', points='outliers', color= 'smoker')
fig.show()
```

```python
fig= px.box(df_t, x= 'time', y='total_bill', points='outliers', color= 'smoker',
           notched= True, title= 'Box Plot of Total bill', hover_data= ['day'])
fig.show()
```

```python
fig= px.strip(df_t, x= 'day', y='tip')
fig.show()
```
